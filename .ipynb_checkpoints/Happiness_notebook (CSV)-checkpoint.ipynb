```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Define the list of countries to analyze
countries = ['Albania', 'Argentina', 'Australia', 'Austria', 'Benin', 'Bolivia', 'Bosnia and Herzegovina', 'Brazil', 'Bulgaria', 'Burkina Faso', 'Cameroon', 'Canada', 'Chile', 'Colombia', 'Congo (Brazzaville)', 'Costa Rica', 'Croatia', 'Denmark', 'Dominican Republic', 'Ecuador', 'El Salvador', 'Estonia', 'Finland', 'France', 'Gabon', 'Georgia', 'Germany', 'Ghana', 'Greece', 'Hungary', 'India', 'Indonesia', 'Ireland', 'Israel', 'Italy', 'Ivory Coast', 'Japan', 'Kazakhstan', 'Kenya', 'Kyrgyzstan', 'Latvia', 'Lebanon', 'Lithuania', 'Mali', 'Mauritius', 'Mexico', 'Moldova', 'Mongolia', 'Myanmar', 'Namibia', 'Nepal', 'Netherlands', 'New Zealand', 'Nicaragua', 'Nigeria', 'North Macedonia', 'Norway', 'Peru', 'Philippines', 'Poland', 'Portugal', 'Romania', 'Russia', 'Senegal', 'Serbia', 'Slovakia', 'Slovenia', 'South Africa', 'South Korea', 'Spain', 'Sri Lanka', 'Sweden', 'Switzerland', 'Tanzania', 'Thailand', 'Tunisia', 'Turkey', 'Uganda', 'Ukraine', 'United Kingdom', 'United States', 'Uruguay', 'Uzbekistan', 'Zambia', 'Zimbabwe']

# Create an empty dictionary to store the sentiment scores for each country
sentiment_scores = {}

# Initialize the sentiment analyzer
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()

# Loop through each country in the list and scrape news articles related to that country
for country in countries:
    url = f'https://news.google.com/search?q={country}&hl=en-US&gl=US&ceid=US%3Aen'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    articles = soup.find_all('div', class_='NiLAwe')
    article_texts = [article.text for article in articles]

    # Calculate the sentiment score for the articles related to the country
    sentiment_score = 0
    for text in article_texts:
        sentiment = analyzer.polarity_scores(text)
        sentiment_score += sentiment['compound']
    sentiment_score /= len(article_texts)

    # Add the sentiment score to the dictionary
    sentiment_scores[country] = sentiment_score

# Create a DataFrame from the sentiment scores dictionary and print the results
df_nlp = pd.DataFrame.from_dict(sentiment_scores, orient='index', columns=['Sentiment Score'])
df_nlp = df_nlp.sort_values(by='Sentiment Score', ascending=False)
print(df_nlp)
```
